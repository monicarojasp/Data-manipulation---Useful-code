Formato ISO de fecha es un estándar internacional que establece una notación numérica para representar fechas y horas de forma universal. El formato ISO de fecha es AAAA-MM-DD, donde YYYY representa el año, MM el mes y DD el día del mes. Por ejemplo, la fecha "3 de abril de 2002" en formato ISO se escribe como "2002-04-03".

## <font color=#003d5c>_.strptime()_ </font>
### Este método nos permite leer una cadena de texto con una fecha en un formato indeterminado como un objeto de clase `.datetime`

|Formatos|Descripción|
|-------|-----------|
|`%a`|Nombre del día de la semana abreviado|
|`%A`|Nombre del día de la semana completo|
|`%w`|Día de la semana como número decimal|
|`%d`|Día del mes como número de dos dígitos|
|`%-d`|Día del mes como número decimal|
|`%b`|Nombre del mes abreviado|
|`%B`|Nombre del mes completo|
|`%m`|Mes del año número de dos dígitos|
|`%-m`|Mes del año como número decimal|
|`%y`|Año sin centenas como número de dos dígitos|
|`%-y`|Año sin centenas como número decimal|
|`%Y`|Año completo|
|`%H`|Hora en 24 hrs. con dos dígitos|
|`%-H`|Hora en 24 hrs. como número decimal|
|`%I`|Hora en 12 hrs. con dos dígitos|
|`%-I`|Hora en 12 hrs. como número decimal|
|`%M`|Minutos con dos dígitos|
|`%-M`|Minutos como número decimal|
|`%S`|Segundos con dos dígitos|
|`%-S`|Segundos como número decimal|



```python
fecha = '22 April, 2024 13:20:13'

d1 = datetime.datetime.strptime(fecha, '%d %B, %Y %H:%M:%S')
print(d1)
print(type(d1))
```

## <font color=#003d5c>_datetime.timedelta_ </font>
### La clase timedelta nos permite operar entre fechas.

```python
d1 = datetime.datetime.now()
print("Fecha de hoy:", d1)

#Obtener la fecha de pasado mañana 

d2 = d1 + datetime.timedelta(days=2)

print('Fecha: 2 días a partir de hoy: ', d2)

#Dos semanas a partir de hoy

d3 = d1 + datetime.timedelta(weeks=2)
print('Fecha: 2 semanas a partir de hoy: ', d3)

print('----------------------------------------')
#Obtener una fecha anterior

#Obtener la fecha de pasado mañana 

d4 = d1 - datetime.timedelta(days=2)

print('Fecha: 2 días antes de hoy: ', d4)

#Dos semanas a partir de hoy

d5 = d1 - datetime.timedelta(weeks=2)
print('Fecha: 2 semanas antes de hoy: ', d5)
```

### La fecha del día siguiente

unit: str, optional
Denotes the unit of the arg for numeric arg. Defaults to "ns".

Possible values:

‘W’

‘D’ / ‘days’ / ‘day’

‘hours’ / ‘hour’ / ‘hr’ / ‘h’ / ‘H’

‘m’ / ‘minute’ / ‘min’ / ‘minutes’ / ‘T’

‘s’ / ‘seconds’ / ‘sec’ / ‘second’ / ‘S’

‘ms’ / ‘milliseconds’ / ‘millisecond’ / ‘milli’ / ‘millis’ / ‘L’

‘us’ / ‘microseconds’ / ‘microsecond’ / ‘micro’ / ‘micros’ / ‘U’

‘ns’ / ‘nanoseconds’ / ‘nano’ / ‘nanos’ / ‘nanosecond’ / ‘N’

```python
print(fecha_hoy)
#Contar --> 4
#unit --> argumento que tiene la unidad de tiempo 'D' --. Días
print('La fecha dentro de 4 días será:', fecha_hoy + pd.to_timedelta(4, unit='D'))
print('La fecha dentro de 4 semanas será:', fecha_hoy + pd.to_timedelta(4, unit='W'))
```

## <font color=#003d5c>_.date_range()_ </font>
### Este método nos permite generar un rango de fechas.

```python
#start --> cadena de texto fecha de inicio 
#end ---> cadena de texto fecha de final 
#freq --> Indica la frecuencia de tiempo 
fechas_inicio = pd.date_range(start='24/4/2022', end='24/5/2022', freq='D')
fechas_inicio
```

```python
fechas_final = pd.date_range(start='24/5/2022', end='24/6/2022', freq='D')
fechas_final
```

```python
#Fecha de inicio --> inicia la campaña de ventas de un producto  
#Fecha de final --> cuando finaliza la campaña de ventas del producto

lista = []

for i in range(15):
    lista.append(np.random.randint(2))

print(lista)
#Valor 0 estados producto, categoria 0
#Valor 1 estados producto, categoria 1

df = pd.DataFrame()
df['Inicio_campaña'] = fechas_inicio[:15]
df['Final_campaña'] = fechas_final[:15]
df['Producto'] = lista

df
```

```python
df['Día_inicio'] = df['Inicio_campaña'].dt.day
df['Mes_inicio'] = df['Inicio_campaña'].dt.month
df['Año_inicio'] = df['Inicio_campaña'].dt.year
df['Hora_inicio'] = df['Inicio_campaña'].dt.hour
df['Minutos_inicio'] = df['Inicio_campaña'].dt.minute
df['Segundos_inicio'] = df['Inicio_campaña'].dt.second

df['Semana_día_inicio'] = df['Inicio_campaña'].dt.weekday
df['Duración_campaña'] = df['Final_campaña'] - df['Inicio_campaña']

df
```

```python
df.columns = ['Inicio_campaña', 'Fin_campaña', 'Target','Día de inicio','Mes de inicio','Año de inicio','Hora_inicio',
              'Minuto_inicio','Segundo_inicio','Día de la semana del inicio','Duración']
```

```python
df
```

```python
df['Día_de_duración'] = df['Duración']/datetime.timedelta(days=1)
df['Minutos_de_duración'] = df['Duración']/datetime.timedelta(minutes=1)
df['Segundos_de_duración'] = df['Duración']/datetime.timedelta(seconds=1)

df
```

**Timestamp- DatetimeIndex**

```python
fechas = pd.date_range(start='2022-01-01',end='2022-01-10',freq='D')
datos = [10, 20, 15, 25, 30, 22, 18, 27, 29, 35]
serie_temporal = pd.Series(datos, index=fechas)
print(serie_temporal)

#Acceder a datos utilizando fechas 
dato_1 = serie_temporal['2022-01-05']
rango_fechas = serie_temporal['2022-01-03':'2022-01-08']

#Resampling y frecuencias temporales 
resample_semana = serie_temporal.resample('W').mean()
resample_semana

#operación 

dias_trancurridos = serie_temporal.index.max() - serie_temporal.index.min()
print(dias_trancurridos)
```

```python
#desplazar las fechas en un índice
fechas_desplazadas = serie_temporal.index + pd.to_timedelta(5, unit='d')
fechas_desplazadas

#Calculo de frecuencia mensual 
conteo_mensual = serie_temporal.resample('ME').count()
conteo_mensual
```

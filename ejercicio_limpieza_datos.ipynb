```python
#=Unir y fusionar dataframes

df_proveedores = df[['PROVEEDORES', 'TELEFONO']] 
df_saldos = df[['PROVEEDORES', 'CONTACTO_COMERCIAL', 'SALDO_PENDIENTE']]

df_merged_proveedores = pd.merge(df_proveedores,df_saldos, on='ID')
df_merged_proveedores.head(10)

```

```python
#outer join --> SELECIONA TODOS LOS ELEMENTOS ENTRE AMBOS DATAFRAME COMUNES Y NO COMUNES
#how --> Tipo de join  
#on --> columna de cruce

df_merged_proveedores_outer_join = pd.merge(df_proveedores,df_saldos, how='outer' ,on='ID')
df_merged_proveedores_outer_join
```

```python
#Left join --> nión de los elementos comúnes y todos los elementos comúnes de la izq o primer dataframe
#how --> Tipo de join  
#on --> columna de cruce

df_merged_proveedores_left_join = pd.merge(df_proveedores,df_saldos, how='left' ,on='ID')
df_merged_proveedores_left_join
```

```python
#Right Join unión de los elementos comúnes y todos los elementos comúnes de la derecha o segundo dataframe
#how --> Tipo de join  
#on --> columna de cruce
df_merged_proveedores_right_join = pd.merge(df_proveedores,df_saldos, how='right' ,on='ID')
df_merged_proveedores_right_join
```

```python
#Join
#lsuffix -->
#rsuffix --> 
df_join = df.join(df_merged_proveedores, rsuffix='_der')
df_join.head()
```

```python
#Conatenación
#axis =0 --> filas 
#axis = 1 --> columnas
df_concatenar = pd.concat([df_proveedores, df_saldo], axis=0)

df_concatenar2 = pd.concat([df_proveedores, df_saldo], keys=['PROVEEDOR', 'SALDO'], axis=0)
df_concatenar2
```

```python
columna = ['PROVEEDORES', 'CONTACTO_COMERCIAL', 'EMAIL', 'TELEFONO', 'SALDO_PENDIENTE', 'FECHA_ULTIMA_COMPRA' ]
df = pd.read_excel(path_data, 
                   header=0,
                   sheet_name='proveedores',
                   index_col='ID',
                   parse_dates=['Fecha de última compra']
                    )
df.columns = columna 
df
```

```python
#Manejo de fechas
df['AÑO'] = df['FECHA_ULTIMA_COMPRA'].dt.year
df['MES'] = df['FECHA_ULTIMA_COMPRA'].dt.month
df['MES_NOMBRE'] = df['FECHA_ULTIMA_COMPRA'].dt.month_name()
df['DÍA'] = df['FECHA_ULTIMA_COMPRA'].dt.day
df['ENTRE_SEMANA'] = df['FECHA_ULTIMA_COMPRA'].dt.weekday
#Calcular los días desde hoy hasta la fecha de la columna
df['DÍAS_DESDE'] = (pd.to_datetime('today') - df['FECHA_ULTIMA_COMPRA']).dt.days
df.head()
```

```python
#Renombre de columnas
#columns={'nombre_viejo' : 'nuevo_nombre'}, inplace = True "Cambio en el Dataframe original"
df.rename(columns={'AÑO': 'AÑO_ULTIMA_COMPRA'}, inplace=True)
df.head()
```
